```python
import pandas as pd 
path="C:\\Users\\hp\\OneDrive\\Documents\\Custom Office Templates\\Desktop\\ml dataset\\Data\\homeprices.csv"

data=pd.read_csv(path)
print(data)
print(data.info())

print("rows:",len(data))
print("columns:",len(data.columns))
print(data.shape)

inputs=data.drop('price',axis=1)
outputs=data['price']

print(inputs)
print("----output----")

print(outputs)


#data visulization 
import matplotlib.pyplot as plt
plt.scatter(inputs,outputs,marker='*',s=100,c='green')
plt.title("house area vs price")
plt.xlabel("area")
plt.ylabel("price")
plt.show()

import sklearn
from sklearn import linear_model

## create linear regression  model
model=linear_model.LinearRegression()
model.fit(inputs,outputs) #train model
result= model.predict([[2600]])
print("predicted output:",result)

#r square===> it shows how much of the change in the output is explained by your model (how well model fits the data )
r2=model.score(inputs,outputs)
print("r2 value:",r2*100)
m=model.coef_
c=model.intercept_
print("slope:",m)
print("intercept:",c)
y=m*2600+c
print("manual calculation:",y)

plt.scatter(inputs,outputs,marker='*',s=100,c='green')
plt.plot(inputs.values,model.predict(inputs),color="red",label='regression line')
plt.title("house area vs price")
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

# 1. Load Data
path = r"C:\Users\hp\OneDrive\Documents\Custom Office Templates\Desktop\ml dataset\Data\Salary_dataset_LR.csv"
data = pd.read_csv(path)

# 2. Clean and Split Data
# Often these CSVs have an 'Unnamed: 0' column which is just a serial number.
# We only want 'YearsExperience' as our input.
if 'Unnamed: 0' in data.columns:
    data = data.drop('Unnamed: 0', axis=1)

# Now 'inputs' will have exactly 1 column: YearsExperience
inputs = data.drop('Salary', axis=1)
outputs = data['Salary']

# 3. Create and Train Model
model = linear_model.LinearRegression()
model.fit(inputs, outputs)

# 4. Predict
test_val = 5 
# Now that 'inputs' has only 1 column, this DataFrame matching will work!
test_df = pd.DataFrame([[test_val]], columns=inputs.columns)
result = model.predict(test_df)

# 5. Metrics
r2 = model.score(inputs, outputs)
m = model.coef_[0]
c = model.intercept_

print(f"Predicted Salary for {test_val} years: {result[0]:.2f}")
print(f"R2 Score: {r2*100:.2f}%")
print(f"Slope (m): {m:.2f}")
print(f"Intercept (c): {c:.2f}")

# 6. Visualization
plt.figure(figsize=(10, 6))
# Scatter plot of actual data
plt.scatter(inputs, outputs, marker='*', s=100, c='green', label='Actual Data')

# Regression Line
plt.plot(inputs, model.predict(inputs), color="red", linewidth=2, label='Regression Line')

plt.title("Salary Prediction: Experience vs Salary", fontsize=14)
plt.xlabel("Years of Experience", fontsize=12)
plt.ylabel("Salary", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
