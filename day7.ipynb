# lambda function

```python
##lambda is a way to create small, anonymous function**>online function
```

```python
def add(x,y):
    return x+y
```

```python
add= lambda x,y:x+y
print(add(3,5))
      
```

```python
it is usefull when you want to pass a quick function into another function for example sorting ,mappingand filtering data
```

```python
#map()
num=[1,2,3,4]
square=list(map(lambda x:x**2,num))
print(square)
```

```python
#filter 
num=[1,2,3,4,5,6]
evens=list(filter(lambda x:x%2==0,num))
print(evens)
```

```python
#sorted
words=["pavan","pradeep","sagar","madan"]
print(sorted(words,key=lambda x:x[-0]))
```
