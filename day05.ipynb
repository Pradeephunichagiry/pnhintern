```python
if op == '+':
    add = a + b
    print("Sum:", add)
```

```python
elif op == '-':
    sub = a - b
    print("sub:", sub)
```

```python
elif op == '*':
    mul = a * b
    print("Mul:", mul)
```

```python
elif op == '/':
    if b == 0:
        print("Dont enter denominator zero")
    else:
        div = a / b
        print("div:", div)
```

```python
else:
    print("Invalid operator")
```

```python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
op = input("Enter the operator (+ - * /): ")

if op == '+':
    print("Sum =", a + b)

elif op == '-':
    print("Subtraction =", a - b)

elif op == '*':
    print("Multiplication =", a * b)

elif op == '/':
    if b == 0:
        print("Denominator cannot be zero")
    else:
        print("Division =", a / b)

else:
    print("Invalid operator")
```
