```python
# polynomial data visualization
import pandas as pd 
path="C:\\Users\\hp\\OneDrive\\Documents\\Custom Office Templates\\Desktop\\ml dataset\\Data\\dataset.csv"
data=pd.read_csv(path)
print(data)
print(data.info())

inputs=data.drop("Salary",axis=1)
outputs=data.drop("Level",axis=1)
print(inputs)
print(outputs)

import matplotlib.pyplot as plt
plt.scatter(inputs,outputs,color="red")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show

import sklearn
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(inputs,outputs)


import matplotlib.pyplot as plt
plt.scatter(inputs,outputs,color="red")
plt.scatter(inputs,model.predict(inputs),label="Regression Line")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend()
plt.show

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

modelPF=PolynomialFeatures(degree=7)
poly=modelPF.fit_transform(inputs)

modelLR=LinearRegression()
modelLR.fit(poly,outputs)

plt.scatter(inputs,outputs,color="red",label="Original data")
plt.scatter(inputs,modelLR.predict(poly),label="Polynomial regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend()
plt.show()

print("r2 value using linear regression :", model.score(inputs,outputs))
print("r2 value using polynomial regression :",modelLR.score(poly,outputs))

l=5
Salary_pred=modelLR.predict(modelPF.transform([[1]]))
print( "Predicted Salary :",Salary_pred)

level=float(input("enter Level:"))
pred=modelLR.predict(modelPF.transform([[level]]))
print( "Predicted Salary for ui :",pred)

```
