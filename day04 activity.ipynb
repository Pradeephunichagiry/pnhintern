```python
###leap year
year = int(input("Enter a year: "))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
## vowel or constent
char = input("Enter a character: ").lower()
if char.isalpha() and len(char) == 1:
    if char in 'aeiou':
        print(f"{char} is a vowel.")
    else:
        print(f"{char} is a consonant.")
else:
    print("Please enter a single valid alphabet letter.")

```

```python
##grade calculator
perc = float(input("Enter your percentage: "))
if perc >= 90:
    print("Grade: A+")
elif perc >= 80:
    print("Grade: A")
elif perc >= 70:
    print("Grade: B")
elif perc >= 60:
    print("Grade: C")
elif perc >= 50:
    print("Grade: D")
else:
    print("Grade: Fail")
```

```python
##palindrome
text = input("Enter a  number: ")
if text == text[::-1]:
    print("It is a palindrome.")
else:
    print("It is not a palindrome.")
```

```python
## divisiable by 5 &11
num = int(input("Enter a number: "))
if num % 5 == 0 and num % 11 == 0:
    print(f"{num} is divisible by both 5 and 11.")
else:
    print(f"{num} is not divisible by both.")
```

```python
##electricity bill
units = float(input("Units: "))
if units <= 50: bill = units * 0.50
elif units <= 150: bill = 25 + (units-50) * 0.75
else: bill = 100 + (units-150) * 1.20
print(f"Total: {bill + bill*0.20}") 
```

```python
##prime number
num = int(input("Enter number: "))
if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
## charchter checking
ch = input("Enter a character: ")

if ch.isalpha():
    print(f"{ch} is an Alphabet.")
elif ch.isdigit():
    print(f"{ch} is a Digit.")
else:
    print(f"{ch} is a Special Character.")
```

```python
##Quadratic Roots
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
d = (b**2) - (4*a*c)

if d > 0:
    root1 = (-b + d**0.5) / (2*a)
    root2 = (-b - d**0.5) / (2*a)
    print("Two roots:", root1, "and", root2)
elif d == 0:
    root = -b / (2*a)
    print("One root:", root)
else:
    print("No real roots")
```

```python
####employee bonus
salary = float(input("Enter salary: "))
years = int(input("Enter years of service: "))
if years > 5:
    bonus = salary * 0.05
    print(f"Bonus is: {bonus}")
else:
    print("No bonus for service less than 5 years.")
```

```python
##type of triangle
s1 = int(input("Side 1: "))
s2 = int(input("Side 2: "))
s3 = int(input("Side 3: "))
if s1 == s2 == s3:
    print("Equilateral Triangle")
elif s1 == s2 or s2 == s3 or s1 == s3:
    print("Isosceles Triangle")
else:
    print("Scalene Triangle")
```

```python
### eligibility check
held = int(input("Number of classes held: "))
attended = int(input("Number of classes attended: "))
percentage = (attended / held) * 100
print(f"Attendance: {percentage}%")

if percentage >= 75:
    print("You are eligible for the exam.")
else:
    print("Not eligible due to low attendance.")
```
