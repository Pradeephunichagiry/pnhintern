```python
users_pnh = {
    "admin": "1234",
    "pnh": "pnh84312",
    "pradeep": "pnh_013"
}
def login():
    print("--- Welcome to the Login Page ---")
    username = input("Enter Username: ")
    if username in users_pnh:
        password = input("Enter Password: ")
        if password == users_pnh[username]:
            print("Login Successful! Welcome back.")
        else:
            print("Invalid Password ")  
    else:
        print("Invalid Username ")
login()
```

```python

   users_pnh = {
    "admin": "1234",
    "pnh": "pnh84312",
    "pradeep": "pnh_013"
}
def register():
    print("\n--- Create a New Account ---")
    new_user = input("Choose a Username: ")
    
    if new_user in users_pnh:
        print("Error: Username already exists. Try a different one.")
    else:
        new_pass = input("Choose a Password: ")
        # Adding the new user to our dictionary
        users_pnh[new_user] = new_pass
        print("Registration Successful! You can now log in.")
def login():
    print("\n--- Login to Your Account ---")
    username = input("Enter Username: ")
    
    if username in users_pnh:
        password = input("Enter Password: ")
        if password == users_pnh[username]:
            print(f"Login Successful! Welcome back, {username}.")
        else:
            print("Invalid Password.")  
    else:
        print("Invalid Username.")
def main():
    while True:
        print("\n--- Welcome System ---")
        print("1. Register")
        print("2. Login")
        print("3. Exit")
        choice = input("Select an option (1-3): ")

        if choice == '1':
            register()
        elif choice == '2':
            login()
        elif choice == '3':
            print("Thankyou!")
            break
        else:
            print("Invalid choice, please try again.")
main()
```
