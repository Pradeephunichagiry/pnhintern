## arthematic opertions

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Add:", a + b)
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Sub:", a - b)
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Mul:", a * b)
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Div:", a // b)
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("modulus:", a % b)
```

```python
##average of three numbers
a = int(input("Number 1: "))
b = int(input("Number 2: "))
c = int(input("Number 3: "))
sum_val = a + b + c
print("Average:", sum_val / 3)
```

```python
##Rectangle Area and Perimeter
l = int(input("Length: "))
w = int(input("Width: "))
print("Area:", l * w)
print("Perimeter:", 2 * (l + w))
```

```python
##simple intrest
p = int(input("Principal: "))
r = int(input("Rate: "))
t = int(input("Time: "))
intrest = (p * r * t) / 100
print("Simple Interest:", intrest)
```

```python
###calculate the square and cube
n = int(input("Enter number: "))
print("Square:", n ** 2)
print("Cube:", n ** 3)
```

```python
##remainder and quotient of two numbers.
a = int(input("Enter a: "))
b = int(input("Enter b: "))
print("Quotient:", a // b)
print("Remainder:", a % b)
```

```python
##converting temperature from Celsius to Fahrenheit
c = int(input("Enter Celsius: "))
f = (c * 9/5) + 32
print("Fahrenheit:", f)
```

```python
###calculate the total marks, average, and percentage of a student
m1 = float(input("Marks 1: "))
m2 = float(input("Marks 2: "))
m3 = float(input("Marks 3: "))
total = m1 + m2 + m3
print("Total:", total)
print("Percentage:", (total / 300) * 100)
```

```python
##Area of a Circle
r = int(input("Radius: "))
print("Area:", 3.14 * r * r)
```

```python
###Profit or Loss
cp = float(input("Cost Price: "))
sp = float(input("Selling Price: "))
profit = max(0, sp - cp)
loss = max(0, cp - sp)
print("Profit =", profit)
print("Loss =", loss)
```

```python
##Distance Travelled
s = float(input("Speed: "))
t = float(input("Time: "))
print("Distance:", s * t)
```

```python
##Power of a Number
base = int(input("Base: "))
pwr = int(input("Power: "))
print("Result:", base ** pwr)
```

```python
##Swap Two Numbers
a = int(input("a: "))
b = int(input("b: "))
a = a + b
b = a - b
a = a - b
print("Swapped a:", a, "b:", b)
```

```python
###Gross Salary
basic = float(input("Basic: "))
hra = float(input("HRA: "))
da = float(input("DA: "))
print("Gross Salary:", basic + hra + da)
```

```python
##arithmetic operator precedence
result = 10 + 2 * 3
print(result)
```
