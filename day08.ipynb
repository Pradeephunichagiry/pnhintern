```python
pos = []
neg = []
a = [1, 2, 3, 4, -5, 6, -7, -8, 1, -3, 4, 5, 7]
for i in a:   
    if i > 0:   
        pos.append(i)
    elif i < 0:   
        neg.append(i)
print("pos:", pos)
print("neg:",neg)
```
