```python
import pandas as pd

path = "C:\\Users\\hp\\OneDrive\\Documents\\Custom Office Templates\\Desktop\\ml dataset\\Data\\Exammarks.csv"
data = pd.read_csv(path)
print(data)
print(data.info())
print("rows:",len(data))
print("columns:",len(data.columns))
print(data.shape)
medianvalue=data.hours.median()
print(medianvalue)

data.hours=data.hours.fillna(medianvalue)
print(data)

inputs=data.drop("marks",axis=1)
outputs=data[['marks']]
print(inputs)
print(outputs)

import matplotlib.pyplot as plt
plt.figure(figsize=(18,5))
plt.subplot(1,3,1)
plt.scatter(data.hours,data.marks,marker="*",color='black')
plt.xlabel("hours studied")
plt.ylabel("Marks obtained")
plt.title("Hours vs Marks")
plt.grid(True)

plt.subplot(1,3,2)
plt.scatter(data.age,data.marks,marker="o",color='red')
plt.xlabel("age")
plt.ylabel("Marks obtained")
plt.title("age vs Marks")
plt.grid(True)


plt.subplot(1,3,3)
plt.scatter(data.internet,data.marks,marker="o",color='red')
plt.xlabel("internet")
plt.ylabel("Marks obtained")
plt.title("internet vs Marks")
plt.grid(True)

plt.show()

import sklearn
from sklearn import linear_model

## create linear regression  model
model=linear_model.LinearRegression()
model.fit(inputs,outputs) #train model
result= model.predict([[6.83,15,1]])
print("predicted output:",result)

Hours=float(input("enter the no.of hours the student studied:"))
age=int(input("enter the age of the student:"))
internet=int(input("wheather students has used internet or not, enter 0 or 1 :"))
pred=model.predict([[Hours,age,internet]])
print("predicted marks:",pred)


m1,m2,m3=model.coef_[0]
intercept=model.intercept_[0]

print("coefficients:")
print(f"Hours:{m1},age:{m2},internet:{m3}")
print("intercept:",intercept)

y=((m1*Hours)+(m2*age)+(m3+internet)+intercept)
print("manual cal:",y)

r2=model.score(inputs,outputs)
print("manual cal:",r2*100)

from sklearn.linear_model import LinearRegression
model_hours = LinearRegression()
model_hours.fit(data[['hours']], data[['marks']])
r2_hours = model_hours.score(data[['hours']], data[['marks']])
print("r2 value using hours only:", r2_hours * 100)

plt.scatter(data.hours,data.marks,marker="*",color='black',label="Original Data")
plt.scatter(Hours,pred,marker="x",color="red",s=200,label="New Input")
plt.xlabel("hours studied")
plt.ylabel("Marks obtained")
plt.title("Hours vs Marks")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression

# Load Dataset
path = "C:\\Users\\hp\\OneDrive\\Documents\\Custom Office Templates\\Desktop\\ml dataset\\Data\\insurance.csv"
data = pd.read_csv(path)

# Clean column names
data.columns = data.columns.str.strip().str.lower()

print(data.head())
print(data.info())
print("rows:", len(data))
print("columns:", len(data.columns))
print(data.shape)

# Fill missing values (if any) using median of age
medianvalue = data['age'].median()
data['age'] = data['age'].fillna(medianvalue)

print(data)

# Select inputs and output
inputs = data[['age', 'bmi', 'children']]
outputs = data[['charges']]

print(inputs.head())
print(outputs.head())

# Visualization
plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
plt.scatter(data['age'], data['charges'], marker="*", color='black')
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Age vs Charges")
plt.grid(True)

plt.subplot(1,3,2)
plt.scatter(data['bmi'], data['charges'], marker="o", color='red')
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.title("BMI vs Charges")
plt.grid(True)

plt.subplot(1,3,3)
plt.scatter(data['children'], data['charges'], marker="o", color='blue')
plt.xlabel("Children")
plt.ylabel("Charges")
plt.title("Children vs Charges")
plt.grid(True)

plt.show()

# Create Linear Regression Model
model = linear_model.LinearRegression()
model.fit(inputs, outputs)

# Example Prediction
result = model.predict([[25, 30.5, 2]])
print("Predicted output for sample [25,30.5,2]:", result)

# User Input Prediction
age = int(input("Enter age: "))
bmi = float(input("Enter BMI: "))
children = int(input("Enter number of children: "))

pred = model.predict([[age, bmi, children]])
print("Predicted insurance charges:", pred)

# Coefficients
m1, m2, m3 = model.coef_[0]
intercept = model.intercept_[0]

print("coefficients:")
print(f"age: {m1}, bmi: {m2}, children: {m3}")
print("intercept:", intercept)

# Manual Calculation
y = (m1*age) + (m2*bmi) + (m3*children) + intercept
print("Manual calculation:", y)

# R2 Score
r2 = model.score(inputs, outputs)
print("R2 value:", r2 * 100)

# Single Variable Model (Age Only)
model_age = LinearRegression()
model_age.fit(data[['age']], data[['charges']])
r2_age = model_age.score(data[['age']], data[['charges']])
print("R2 value using age only:", r2_age * 100)

# Final Plot with Prediction Point
plt.scatter(data['age'], data['charges'], marker="*", color='black', label="Original Data")
plt.scatter(age, pred[0][0], marker="x", color="red", s=200, label="New Input")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Age vs Charges")
plt.grid(True)
plt.legend()
plt.show()
```
